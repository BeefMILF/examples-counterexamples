```python
import numpy as np
from sklearn.datasets import make_blobs

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x1 = np.array([10, 0])
x2 = np.array([-10, 0])

xs = np.vstack((x1, x2))
```

```python
dataset, labels = make_blobs(n_samples=100, centers=xs)
dataset = dataset @ np.array([[1, 0], [0, 100]])
```

```python
plt.scatter(dataset[:, 0], dataset[:, 1], c=labels)
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=1)
reduced_dataset = pca.fit_transform(dataset)
inverse_transformed_dataset = pca.inverse_transform(reduced_dataset)
```

```python
plt.scatter(inverse_transformed_dataset[:, 0], inverse_transformed_dataset[:, 1], c=labels)
plt.show()
```
