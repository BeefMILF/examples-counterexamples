```python
import numpy as np

from sklearn.datasets import load_sample_image
from skimage.util.shape import view_as_blocks

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def show_image(img, cmap='gray'):
  plt.imshow(img, cmap=cmap)
  plt.axis('off')
  plt.show()
```

Load images and convert them to grayscale

```python
flower_image = load_sample_image('flower.jpg').mean(axis=-1)
china_image = load_sample_image('china.jpg').mean(axis=-1)
```

```python
print('flower image shape: {}'.format(flower_image.shape))
show_image(flower_image)
```

```python
block_shape = 61, 40 
flower_blocks = view_as_blocks(flower_image, block_shape)
flower_blocks.shape
```

To retrieve image from first row, get it and use `np.hstack`

```python
flower_first_row = flower_blocks[1]

show_image(np.hstack(flower_first_row))
```

`hstack` squashes `ndarray`'s zeroth and 2nd axis.

It does it by concatenating along the 2nd axis.

```python
hstacked = np.hstack(flower_blocks)
print('block matrix shape:', flower_blocks.shape)
print('shape after hstack:', hstacked.shape)
show_image(hstacked[5])
```

Doing `hstack` two times retrieves original image.

```python
retrieved_flower = np.hstack(np.hstack(flower_blocks))
show_image(retrieved_flower)
```
