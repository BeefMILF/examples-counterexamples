```python
import numpy as np
from scipy.io import wavfile

import matplotlib.pyplot as plt
%matplotlib inline

from IPython.display import Audio
```

# Generating noise

### White noise

```python
white_noise = np.random.randn(50000)
plt.plot(range(white_noise.size), white_noise)
Audio(data=white_noise, rate=14440)
```

### Brownian noise

```python
brown_noise = np.cumsum(np.random.randn(50000))
plt.plot(range(brown_noise.size), brown_noise)
Audio(data=brown_noise, rate=14440)
```

# Generating noise with given Fourier transform

$\hat{f}$ - random binary vector

```python
random_binary = np.random.binomial(2, p=0.5, size=50000)
x_b = np.fft.ifft(random_binary)
plt.plot(range(x_b.size), np.real(x_b))
Audio(data=np.real(x_b), rate=14440)
```

$\hat{f}$ - random normal vector

```python
random_continuous = np.random.randn(50000)
x = np.fft.ifft(random_continuous)
plt.plot(range(x.size), np.real(x))
Audio(data=np.real(x), rate=14440)
```

```python
x_b_corr = np.correlate(x_b, x_b, mode='full')[:x_b.size // 2]
x_b_sd = np.fft.fft(x_b_corr)
```
